````python
def tracklet(vid_name, pred, f_s, f_e):
    if os.path.isfile('/DATA/VRC_STGCN/VRG/STGCN/VideoVRD/vidvrd-dataset/test/{}.json'.format(vid_name)):
        cur_gt = json.load(open('/DATA/VRC_STGCN/VRG/STGCN/VideoVRD/vidvrd-dataset/test/{}.json'.format(vid_name)))
    elif os.path.isfile('/DATA/VRC_STGCN/VRG/STGCN/VideoVRD/vidvrd-dataset/train/{}.json'.format(vid_name)):
        cur_gt = json.load(open('/DATA/VRC_STGCN/VRG/STGCN/VideoVRD/vidvrd-dataset/train/{}.json'.format(vid_name)))
    else:
        print("file not found")
    
    final_traj = {}
    sub = pred.split('-')[0]
    obj = pred.split('-')[2]

    sub_tid = []
    obj_tid = []
    obj_b = {}
    sub_b = {}

    for ele in cur_gt['subject/objects']:
        if ele['category'] == sub:
            sub_tid.append(ele['tid'])
        if ele['category'] == obj:
            obj_tid.append(ele['tid'])
    # if check != 0:
    #     print("object/subject not found in the file") 
    #print(sub_tid, obj_tid)
    for ele in cur_gt['relation_instances']:
        #print(ele['predicate'])
        if ele['predicate'] == pred.split('-')[1] and (ele['subject_tid'] in sub_tid) and (ele['object_tid'] in obj_tid):
            f_s = ele['begin_fid']
            f_e = ele['end_fid']
            print(f'relation {pred} found in video {vid_name} with duration {f_s} to {f_e}')
        
    
    
            #print(f_s, f_e )
            # if f_s == 0 and f_e == 0:
            #     return []

                    
            for frame_i, bbox in enumerate(cur_gt['trajectories']):
                

                if frame_i >= f_s and frame_i < f_e:

                    for box in bbox:
                        cur_obj_box = []
                        cur_sub_box = []

                        if box['tid'] in sub_tid:
                            for coord in box['bbox'].keys():
                                cur_sub_box.append(box['bbox'][coord])
                            sub_b[str(frame_i)] = cur_sub_box
                        if box['tid'] in obj_tid:
                            for coord in box['bbox'].keys():
                                cur_obj_box.append(box['bbox'][coord])
                            obj_b[str(frame_i)] = cur_obj_box
            break
                

    final_traj["sub"] = sub_b
    final_traj["obj"] = obj_b

    return [final_traj]
````

````python
def tracklet_stgcn(vid_name, pred):
    if os.path.isfile('/DATA/VRC_STGCN/VRG/STGCN/VideoVRD/vidvrd-dataset/test/{}.json'.format(vid_name)):
        cur_gt = json.load(open('/DATA/VRC_STGCN/VRG/STGCN/VideoVRD/vidvrd-dataset/test/{}.json'.format(vid_name)))
    elif os.path.isfile('/DATA/VRC_STGCN/VRG/STGCN/VideoVRD/vidvrd-dataset/train/{}.json'.format(vid_name)):
        cur_gt = json.load(open('/DATA/VRC_STGCN/VRG/STGCN/VideoVRD/vidvrd-dataset/train/{}.json'.format(vid_name)))
    else:
        print("file not found")
    
    final_traj = {}
    sub = pred.split('-')[0]
    obj = pred.split('-')[2]
    check = 2
    sub_tid = 0
    obj_tid = 0

    for ele in cur_gt['subject/objects']:
        if ele['category'] == sub:
            sub_tid = ele['tid']
            check -= 1
        if ele['category'] == obj:
            obj_tid = ele['tid']
            check -= 1
    # if check != 0:
    #     print("object/subject not found in the file") 
    #print(sub_tid, obj_tid)
    f_s = 0
    f_e = 0
    for ele in cur_gt['relation_instances']:
        #print(ele['predicate'])
        if ele['predicate'] == pred.split('-')[1] and ele['subject_tid'] == sub_tid and ele['object_tid'] == obj_tid:
            f_s = ele['begin_fid']
            f_e = ele['end_fid']
            print(f'relation {pred} found in video {vid_name} with duration {f_s} to {f_e}')
            break
    
    
    #print(f_s, f_e )
    # if f_s == 0 and f_e == 0:
    #     return []
    obj_b = {}
    sub_b = {}
            
    for frame_i, bbox in enumerate(cur_gt['trajectories']):
        

        if frame_i >= f_s and frame_i < f_e:

            for box in bbox:
                cur_obj_box = []
                cur_sub_box = []

                if box['tid'] == sub_tid:
                    for coord in box['bbox'].keys():
                        cur_sub_box.append(box['bbox'][coord])
                    sub_b[str(frame_i)] = cur_sub_box
                if box['tid'] == obj_tid:
                    for coord in box['bbox'].keys():
                        cur_obj_box.append(box['bbox'][coord])
                    obj_b[str(frame_i)] = cur_obj_box
                

    final_traj["sub"] = sub_b
    final_traj["obj"] = obj_b

    return [final_traj]
````
